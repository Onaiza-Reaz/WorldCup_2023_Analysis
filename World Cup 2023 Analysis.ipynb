## Handling Missing Values

```python
columns_with_missing = ['overs', 'mdns', 'econ', 'mins']

# Mode imputation
for column in columns_with_missing:
  df[column].fillna(df[column].mean(),inplace=True)

```

```python
column_missing=['4s','6s', 'sr', 'not_out','wkts']
for missing in column_missing:
     mode_value = df[missing].mode().iloc[0]
     # Mode might return multiple values, so we take the first one
     df[missing].fillna(mode_value, inplace=True)
```

## Converting Float Data Type into Integer

```python
columns_to_convert = ['wkts', 'wicketball_prob', 'runs_per_ball', 'overs',
                      'mdns', 'econ', '4s', '6s', 'sr', 'not_out', 'mins']

# Convert float columns to integers
df[columns_to_convert] = df[columns_to_convert].astype(int)
```

## Teams with their Runs, Wickets, and Batting Bowling

```python
# Group by team and calculate total runs, wickets, and other metrics
team_stats = df.groupby('team').agg({
    'runs': 'sum',
    'wkts': 'sum',
    "bat_or_bowl" : "value_counts"   # Analyze batting/bowling styles
}).unstack().fillna(0)

print(team_stats.head(10))
print(team_stats.tail(10))
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='team', hue="bat_or_bowl", data=df)
plt.title('Team-wise Performance: Batting vs. Bowling')
plt.xlabel('Team')
plt.ylabel('Count')
plt.legend(title='Batting/Bowling')
plt.show()
```

## Player Performance Analysis

```python
# Group by player and calculate total runs and wickets
player_stats = df.groupby(['team', 'player', 'bat_or_bowl']).agg({
    'runs': 'sum',
    'wkts': 'sum',
}).reset_index()

# Identify leading run-scorers and wicket-takers
leading_run_scorers = player_stats[player_stats['bat_or_bowl']
                                   == 'bat'].nlargest(5, 'runs')
leading_wicket_takers = player_stats[player_stats['bat_or_bowl'] == 'bowl']
.nlargest(5, 'wkts')

print("Leading Run-Scorers:")
print(leading_run_scorers)

print("\nLeading Wicket-Takers:")
print(leading_wicket_takers)
```

## Top Impactful Players

```python
player_stats['impact_metric'] = player_stats['runs'] / player_stats['wkts']
impactful_players = player_stats.nlargest(5, 'impact_metric')

print("\nTop Impactful Players:")
print(impactful_players)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize leading run-scorers
plt.figure(figsize=(12, 6))
sns.barplot(x='player', y='runs', hue='team', data=leading_run_scorers)
plt.title('Leading Run-Scorers')
plt.xlabel('Player')
plt.ylabel('total Runs')
plt.show()

# Visualize leading wicket-takers
plt.figure(figsize=(12, 6))
sns.barplot(x='player', y='wkts', hue='team', data=leading_wicket_takers)
plt.title('Leading Wicket-Takers')
plt.xlabel('Players')
plt.ylabel('Total Wicket')
plt.show()
```

## Opposition Ground Analysis

## Runs and Wicket with Opposite Team

```python
# Group by opposition and calculate total runs and wickets
opposition_stats = df.groupby(['opposition', 'bat_or_bowl']).agg({
    'runs': 'sum',
    'wkts': 'sum',
}).reset_index()

# Visualize performance against different oppositions
plt.figure(figsize=(12, 6))
sns.barplot(x='opposition', y='runs', hue='bat_or_bowl', data=opposition_stats)
plt.title('Team Performance Against Different Oppositions')
plt.xlabel('Opposition')
plt.ylabel('Total Runs/Wickets')
plt.show()
```

## Runs and Wicket In Different Ground

```python
# Group by ground and calculate total runs and wickets
ground_stats = df.groupby(['ground', 'bat_or_bowl']).agg({
    'runs': 'sum',
    'wkts': 'sum',
}).reset_index()

# Visualize performance at different grounds
plt.figure(figsize=(12, 6))
sns.barplot(x='ground', y='runs', hue='bat_or_bowl', data=ground_stats)
plt.title('Team Performance at Different Grounds')
plt.xlabel('Ground')
plt.ylabel('Total Runs/Wickets')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Runs per Wicket for Batsmen
## Wickets per Run for Bowlers

```python
# Calculate a performance metric (e.g., runs per wicket for batsmen, wickets per run for bowlers)
team_performance_metric = df.groupby(['team', 'opposition']).agg({
    'runs': 'sum',
    'wkts': 'sum',
}).reset_index()

team_performance_metric['performance_metric'] = team_performance_metric['runs'] / team_performance_metric['wkts']

# Identify teams excelling against specific oppositions
best_teams_against_opposition = team_performance_metric.groupby('opposition').apply(lambda x: x.loc[x['performance_metric'].idxmax()])

print("Teams Excelling Against Specific Oppositions:")
print(best_teams_against_opposition[['team', 'opposition', 'performance_metric']])
```

```python
# Visualize teams excelling against specific oppositions
plt.figure(figsize=(12, 6))
sns.barplot(x='opposition', y='performance_metric', hue='team', data=best_teams_against_opposition)
plt.title('Teams Excelling Against Specific Oppositions')
plt.xlabel('Opposition')
plt.ylabel('Performance Metric (e.g., Runs per Wicket)')
plt.show()
```

## Temporal Analysis

```python
# Convert 'start_date' to datetime format
df['start_date'] = pd.to_datetime(df['start_date'])
```

```python
# Group by start date and calculate total runs and wickets
team_temporal_stats = df.groupby(['start_date', 'bat_or_bowl']).agg({
    'runs': 'sum',
    'wkts': 'sum',
}).reset_index()

# Visualize team performance trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='start_date', y='runs', hue='bat_or_bowl', data=team_temporal_stats)
plt.title('Team Performance Trends Over Time')
plt.xlabel('Start Date')
plt.ylabel('Total Runs/Wickets')
plt.show()
```
